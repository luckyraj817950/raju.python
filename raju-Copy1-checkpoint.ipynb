```python
def cop(a,b):
    return gcd(a,b)==1
    
def gcd(a,b):
    while(b!=0):
        rem=a%b
        a=b
        b=rem
    return a   
    
a=int(input("enter the number"))
for i in range(5,a):
    for j in range(4,i):
        for k in range(3,j):
            if(j*j + k*k==i*i and cop(i,j) and cop(j,k) and cop(i,k)):
                print(k,j,i)
```

```python
def prime(i):
    if (i<2):
         return False
    for j in range(2,i):
        if(i%j==0):
            return False
    return True
        
def sod(i):
    d=[int(i) for i in str(i)]
    return sum(d)
  
def alldig(i):
    while(i!=0):
        d=i%10
        if(not prime(d)):
            return False
        i=i//10
    return True
            
for i in range(100,10000):
    if(prime(i) and prime(sod(i)) and alldig(i)):
        print(i,end=" ")
```
